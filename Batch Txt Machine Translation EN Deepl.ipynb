```python
## translate all the files in the folder to Chinese

import deepl
import re
import os

def translate_srt_file_deepl(srt_file_path, translated_srt_file_path, target_language, auth_key):
    translator = deepl.Translator(auth_key)

    with open(srt_file_path, 'r', encoding='utf-8') as file:
        srt_content = file.readlines()

    translated_srt_content = []

    for line in srt_content:
        if re.match(r'\d+\n', line) or '-->' in line:
            translated_srt_content.append(line)
        elif line.strip():
            result = translator.translate_text(line.strip(), target_lang=target_language)
            translated_line = result.text
            translated_srt_content.append(translated_line + '\n')
        else:
            translated_srt_content.append('\n')

    with open(translated_srt_file_path, 'w', encoding='utf-8') as file:
        file.writelines(translated_srt_content)

def translate_all_srt_in_folder(folder_path, output_folder_path, target_language, auth_key):
    for filename in os.listdir(folder_path):
        if filename.endswith('.srt'):
            srt_file_path = os.path.join(folder_path, filename)
            translated_srt_file_path = os.path.join(output_folder_path, f"translated_{filename}")
            translate_srt_file_deepl(srt_file_path, translated_srt_file_path, target_language, auth_key)
            print(f"Translated '{filename}'")

# Usage example
#auth_key = "your-deepl-api-auth-key"  # Replace with your DeepL API key
#folder_path = "/path/to/your/folder"  # Replace with your folder path containing SRT files
#output_folder_path = "/path/to/your/output/folder"  # Replace with the path where you want to save translated SRT files
#target_language = "EN"  # Replace with your target language code
#translate_all_srt_in_folder(folder_path, output_folder_path, target_language, auth_key)

# Usage example
auth_key = "0d2d3868-5f43-2c6d-7c44-cc700c90ae13:fx"  # Replace with your DeepL API key
folder_path = "D:/OneDrive/Job/TDM/20231128 - MyLand/04"  # Replace with your folder path containing SRT files
output_folder_path = "D:/OneDrive/Job/TDM/20231128 - MyLand/03"  # Replace with the path where you want to save translated SRT files
target_language = "EN-US"  # Replace with your target language code
translate_all_srt_in_folder(folder_path, output_folder_path, target_language, auth_key)
```
